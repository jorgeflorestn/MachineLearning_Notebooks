# Descripcion de funcion costo y sigmoide para regresion logistica

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
z = np.linspace(-10, 10, 101)
gz = sigmoid(z)
```

```python
plt.figure(figsize = (5, 3))
plt.plot(z[:], gz[:], "b-", linewidth = 2, label = "Funcion sigmoide")
plt.xlabel('z', fontsize = 16)
plt.ylabel('gz(z)', fontsize = 16, rotation = 90)
plt.legend(loc = "upper left", fontsize = 11)
plt.grid(True)
plt.show()
```

```python
X = np.array([0.052, 0.081, 0.096, 0.158, 0.242, 0.253, 0.396, 0.427, 0.549, 0.572])
y = np.array([0,0,0,0,1,0,1,1,1,1])
```

```python
plt.figure(figsize = (5, 3))
plt.plot(X[:], y[:], "b.", linewidth = 2)
plt.xlabel('z', fontsize = 16)
plt.ylabel('gz(z)', fontsize = 16, rotation = 90)
#plt.legend(loc = "upper left", fontsize = 11)
plt.grid(True)
plt.show()
```

## Ejemplo de predicci√≥n 

```python
m = X.size

#leer pesos
theta1 = np.arange(-40, 50, 0.15)
# Hipostesis de la regresion lineal
h = np.array([-0.5 + theta * X for theta in theta1])
# Hipotesis para la regresion logistica
h = sigmoid(h)

# Calculo de costos 
J = (1 / (2 * m)) * ((h - y) ** 2).sum(axis = 1)

# Nuevo calculo de costo
J = -(1 / m) * (y * np.log(h) + (1 - y) * np.log(1- h)).sum(axis = 1)
```

```python
plt.figure(figsize = (8, 5))
plt.plot(theta1, J, "b-", linewidth = 2, label = "Funcion costo lineal")
plt.xlabel('z', fontsize = 16)
plt.ylabel('gz(z)', fontsize = 16, rotation = 90)
#plt.legend(loc = "upper left", fontsize = 11)
plt.grid(True)
plt.show()
```

## Descripcion de la nueva funcion costo $ J (\theta)$

```python
h = np.arange(0.01, 1, 0.01)
J = - np.log(h)

plt.figure(figsize = (8, 5))
plt.plot(h, J, "b-", linewidth = 2.5, label = "Funcion costo lineal")
plt.xlabel('z', fontsize = 16)
plt.ylabel('gz(z)', fontsize = 16, rotation = 90)
plt.legend(loc = "upper left", fontsize = 11)
plt.grid(True)
plt.show()
```

```python
h = np.arange(0.01, 1, 0.01)
J = - np.log(1 - h)

plt.figure(figsize = (8, 5))
plt.plot(h, J, "b-", linewidth = 2, label = "Funcion costo lineal")
plt.xlabel('z', fontsize = 16)
plt.ylabel('gz(z)', fontsize = 16, rotation = 90)
plt.xlabel('Probability', fontsize = 12)
plt.legend(loc = "upper left", fontsize = 11)
plt.grid(True)
plt.show()
```

## *Pendiente* derivada del nuevo costo para el algoritmo de descenso de gradiente

```python
theta = np.array([4.3])

X_new = np.linspace(-1, 2, 1000)
y_proba = sigmoid(-0.5 + theta * X_new)
```

```python
decision_Boundary = X_new[y_proba >= 0.5][0]

decision_Boundary
```

```python
plt.figure(figsize = (8, 5))
plt.plot(X[y == 0], y[y == 0], 'bs', markersize = 6)
plt.plot(X[y == 1], y[y == 1], 'g^', markersize = 6)
plt.plot([decision_Boundary, decision_Boundary], [-1, 2], 'k:', linewidth = 2)


plt.plot(X_new, y_proba, 'g-', label = "Clase 1")
plt.plot(X_new, 1 - y_proba, 'b--', label = "Clase 0")
plt.xlabel('X', fontsize = 12)
plt.xlabel('Probability', fontsize = 12)
plt.legend(loc = 'center left', fontsize = 12)
plt.axis([-1, 2, -0.02, 1.02])
plt.show()
```
